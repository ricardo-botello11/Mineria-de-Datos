## Descripición de los datos
**a) ¿Qué tipo de datos son?**

Los datos encontrados vienen en forma de tabla y fueron obtenidos de “Our World in Data” que es un repositorio de github para COVID-19.

**b) Descripción de las Columnas**

+ País: referente al país de donde viene la información sobre la vacuna.
+ Fecha: fecha de la entrada de datos. Es importante mencionar que algunas de las fechas son de las vacunas diarias y otras son del total de vacunas. 
+ Total_V: número total de vacunas puestas.
+ Personas_V:  Cada persona dependiendo del esquema de inmunización va a recibir una o dos dosis de vacuna. Hasta cierto punto, va a haber más cantidad de vacunas que de personas.
+ PV_Totalmente: Cantidad de personas que ya recibieron su dosis completa de vacuna (normalmente son 2). 
+ V_Diarias: Cantidad de vacunas diarias a la fecha por país.
+ Nombre_V: Nombre de vacuna usada en el país.

## Programación

```python
import pandas as pd
import numpy as np

df = pd.read_csv("C:/Users/rbote/Downloads/country_vaccinations.csv")
df
```

Decidimos eliminar las columnas de iso_code, total_vaccination_per_hundred, people_vaccinated_per_hundred, people_fully_vaccinated_per_hundred, daily_vaccinations_per_million, source_name y source_website ya que creemos que no afectarán el análisis que haremos de la base de datos.    

```python
#Función para quitar columnas
#http://exponentis.es/como-eliminar-columnas-de-un-dataframe-de-python-pandas
df1 = df.drop(df.columns[[1,6,8,9,10,11,13,14]], axis = 'columns')
df1
```

Cambiamos el nombre de las columnas a español para facilitarnos el manejo de las base de datos.

```python
#Función para cambio de nombre
#https://www.iteramos.com/pregunta/7805/-renombrar-las-columnas-en-los-pandas-
df1.rename(columns = {'country':'País'}, inplace=True)
df1.rename(columns = {'date':'Fecha'}, inplace=True)
df1.rename(columns = {'total_vaccinations':'Total_V'}, inplace=True)
df1.rename(columns = {'people_vaccinated':'Personas_V'}, inplace=True)
df1.rename(columns = {'people_fully_vaccinated':'PV_Totalmente'}, inplace=True)
df1.rename(columns = {'daily_vaccinations':'V_Diarias'}, inplace=True)
df1.rename(columns = {'vaccines':'Nombre_V'}, inplace=True)
df1
```
